### A3. 이웃한 칸

```python
def solution(board, h, w):
    cnt = 0
    n = len(board)
    color = board[h][w]
    dh = [-1, 1, 0, 0]
    dw = [0, 0, -1, 1]

    for i in range(4):
        h_check, w_check = h + dh[i], w + dw[i]
        if 0 <= h_check < n and 0 <= w_check < n:
            if board[h_check][w_check] == color:
                cnt += 1
    
    return cnt
```

```python
solution([["blue", "red", "orange", "red"], ["red", "red", "blue", "orange"], ["blue", "orange", "red", "red"], ["orange", "orange", "red", "blue"]], 1, 1)
```

```python
solution([["yellow", "green", "blue"], ["blue", "green", "yellow"], ["yellow", "blue", "blue"]], 0, 1)
```

```python
solution([["yellow", "green", "blue"], ["blue", "green", "yellow"], ["yellow", "blue", "blue"]], 2, 1)
```

### A4. 데이터 분석

```python
def solution(data, ext, val_ext, sort_by):
    ext_code = {'code': 0, 'date': 1, 'maximum': 2, 'remain': 3}
    
    if ext == 'code':
        data = [d for d in data if d[ext_code[ext]] < val_ext]
        data.sort(key = lambda x: x[ext_code[sort_by]])

    elif ext == 'date':
        data = [d for d in data if d[ext_code[ext]] < val_ext]
        data.sort(key = lambda x: x[ext_code[sort_by]])

    elif ext == 'maximum':
        data = [d for d in data if d[ext_code[ext]] < val_ext]
        data.sort(key = lambda x: x[ext_code[sort_by]])

    else:
        data = [d for d in data if d[ext_code[ext]] < val_ext]
        data.sort(key = lambda x: x[ext_code[sort_by]])

    return data
```

```python
solution([[1, 20300104, 100, 80], [2, 20300804, 847, 37], [3, 20300401, 10, 8]], 'date', 20300501, 'remain')
```

### A4. 데이터 분석

```python
def solution(data, ext, val_ext, sort_by):
    ext_code = ['code', 'date', 'maximum', 'remain']
    
    data = [d for d in data if d[ext_code.index(ext)] < val_ext]
    data.sort(key = lambda x: x[ext_code.index(sort_by)])
    
    return data
```

```python
solution([[1, 20300104, 100, 80], [2, 20300804, 847, 37], [3, 20300401, 10, 8]], 'date', 20300501, 'remain')
```

### A5. 달리기 경주 - 시간초과 에러 발생

```python
def solution(players, callings):
    for player in callings:
        idx = players.index(player)
        players[idx], players[idx-1] = players[idx-1], players[idx]
    return players
```

```python
solution(["mumu", "soe", "poe", "kai", "mine"], ["kai", "kai", "mine", "mine"])
```

### A5. 달리기 경주 - dictionary 사용

```python
def solution(players, callings):
    rank = {player: idx for idx, player in enumerate(players)}
    for player in callings:
        idx = rank[player]
        rank[player] -= 1
        rank[players[idx-1]] += 1
        players[idx], players[idx-1] = players[idx-1], players[idx]
    return players
```

```python
solution(["mumu", "soe", "poe", "kai", "mine"], ["kai", "kai", "mine", "mine"])
```
