## Q1. 정수를 나선형으로 배치하기 - O


https://school.programmers.co.kr/learn/courses/30/lessons/181832

### A1. 정수를 나선형으로 배치하기

```python
def solution(n):
    if n == 1:
        return [[1]]

    answer = [[0 for j in range(n)] for i in range(n)]

    x, y = 0, 0
    dir = 'to_right'

    for i in range(n*n):
        answer[x][y] = i + 1

        if dir == 'to_right':
            y += 1
            if y == n-1 or answer[x][y+1] != 0:
                dir = 'to_bottom'

        elif dir == 'to_bottom':
            x += 1
            if x == n-1 or answer[x+1][y] != 0:
                dir = 'to_left'

        elif dir == 'to_left':
            y -= 1
            if y == 0 or answer[x][y-1] != 0:
                dir = 'to_top'

        elif dir == 'to_top':
            x -= 1
            if x == n-1 or answer[x-1][y] != 0:
                dir = 'to_right'

    return answer
```

```python
solution(5)
```

## Q2. 주사위 게임 3 - O


https://school.programmers.co.kr/learn/courses/30/lessons/181916

### A2. 주사위 게임 3

```python
def solution(a, b, c, d):
    nums = [a, b, c, d]
    cnts = {a: nums.count(a), 
            b: nums.count(b), 
            c: nums.count(c), 
            d: nums.count(d)}
    cnts_reversed = dict(map(reversed, cnts.items()))

    # 가장 많이 나온 숫자와 그 횟수 추출
    most_cnt = max(cnts.values())
    most_num = cnts_reversed[most_cnt]

    # 가장 적게 나온 숫자와 그 횟수 추출
    fewest_cnt = min(cnts.values())
    fewest_num = cnts_reversed[fewest_cnt]
    
    # 1
    if most_cnt == 4:
        return 1111 * most_num
    # 2
    elif most_cnt == 3:
        return (10*most_num + fewest_num)**2
    elif most_cnt == 2:
        # 3
        if fewest_cnt == 2:
            nums_set = list(set(nums))
            return (nums_set[0] + nums_set[1]) * abs(nums_set[0] - nums_set[1])
        # 4
        else:
            while most_num in nums:
                nums.remove(most_num)
            return nums[0] * nums[1]
    # 5
    else:
        return min(nums)
```

### A2. 주사위 게임 3

```python
def solution(a, b, c, d):
    nums = [a, b, c, d]
    cnts = [nums.count(num) for num in nums]

    # 1
    if max(cnts) == 4:
        return 1111 * a
    # 2
    elif max(cnts) == 3:
        return (10 * nums[cnts.index(3)] + nums[cnts.index(1)])**2
    elif max(cnts) == 2:
        # 3
        if min(cnts) == 1:
            return eval('*'.join([str(nums[idx]) for idx, cnt in enumerate(cnts) if cnt == 1]))
        # 4
        else:
            return (max(nums) + min(nums)) * abs(max(nums) - min(nums))
    # 5
    else:
        return min(nums)
```

## Q3. 배열 조각하기 - O


https://school.programmers.co.kr/learn/courses/30/lessons/181893

### A3. 배열 조각하기

```python
def solution(arr, query):
    for i in range(len(query)):
        if i % 2 == 0 :
            arr = arr[:query[i]+1]
        else :
            arr = arr[query[i]:]
    return arr
```

```python
solution([0, 1, 2, 3, 4, 5], [4, 1, 2])
```

## Q4. 코드 처리하기 - O


https://school.programmers.co.kr/learn/courses/30/lessons/181932

### A4. 코드 처리하기

```python
def solution(code):
    mode = 0
    ret = ''

    for idx in range(len(code)):
        if mode == 0:
            if code[idx] != '1':
                if idx % 2 != 0:
                    continue
                ret += code[idx]
            else:
                mode = 1
        
        elif mode == 1:
            if code[idx] != '1':
                if idx % 2 == 0:
                    continue
                ret += code[idx]
            else:
                mode = 0

    return ret if ret != '' else 'EMPTY'
```

```python
solution('')
```

```python
solution('111')
```

## Q5. 배열 만들기 2 - O

### A5. 배열 만들기 2

```python
def solution(l, r):
    result = []

    for num in range(l, r+1):
        num_list = list(set(str(num)))

        if num_list == ['5'] or num_list == ['5', '0'] or num_list == ['0', '5']:
            result.append(num)
        else:
            continue

    return result if result != [] else [-1]
```

```python
solution(10, 20)
```

```python
solution(1, 105)
```

```python
solution(5, 5)
```

### A5. 배열 만들기 2

```python
def solution(l, r):
    result = []
    for num in range(l, r + 1):
        if not set(str(num)) - set(['0', '5']):
            result.append(num)
    return result if result else [-1]
```

## Q6. 전국 대회 선발 고사 - O


https://school.programmers.co.kr/learn/courses/30/lessons/181851

### A6. 전국 대회 선발 고사

```python
def solution(rank, attendance):
    rank_True = []

    for n in range(0, len(rank)):
        if attendance[n]:
            rank_True.append(rank[n])
        else:
            continue

    rank_True = sorted(rank_True)

    a = rank.index(rank_True[0])
    b = rank.index(rank_True[1])
    c = rank.index(rank_True[2])

    return 10000 * a + 100 * b + c
```

```python
solution([3, 7, 2, 5, 4, 6, 1], [False, True, True, True, True, False, False])
```

### A6. 전국 대회 선발 고사

```python
def solution(rank, attendance):
    rank_attend = []

    for idx, attend in enumerate(attendance):
        if attend:
            rank_attend.append((rank[idx], idx))

    rank_attend.sort()

    a, b, c = rank_attend[:3]

    return 10000 * a[1] + 100 * b[1] + c[1]
```

```python
solution([3, 7, 2, 5, 4, 6, 1], [False, True, True, True, True, False, False])
```

## Q9. 그림 확대 - O


https://school.programmers.co.kr/learn/courses/30/lessons/181836

### A9. 그림 확대

```python
def solution(picture, k):
    result = []
    
    row = len(picture)
    col = len(picture[0])

    for r in range(row):
        el = ''
        for c in range(col):
            el += picture[r][c] * k
        for i in range(k):
            result.append(el)

    return result
```

```python
solution([".xx...xx.", "x..x.x..x", "x...x...x", ".x.....x.", "..x...x..", "...x.x...", "....x...."], 2)
```

```python
solution(["x.x", ".x.", "x.x"], 3)
```

### A9. 그림 확대

```python
def solution(picture, k):
    result = []

    for i in range(len(picture)):
        result += [picture[i].replace('.', '.'*k).replace('x', 'x'*k)] * k

    return result
```

```python
solution([".xx...xx.", "x..x.x..x", "x...x...x", ".x.....x.", "..x...x..", "...x.x...", "....x...."], 2)
```

```python
solution(["x.x", ".x.", "x.x"], 3)
```

## Q10. 정사각형으로 만들기 - O


https://school.programmers.co.kr/learn/courses/30/lessons/181830

### A10. 정사각형으로 만들기

```python
def solution(arr):
    row = len(arr)
    col = len(arr[0])

    if row > col:
        for el in arr:
            for i in range(row - col):
                el.append(0)
        return arr
    elif row < col:
        for i in range(col - row):
            arr.append([0] * col)
        return arr
    else:
        return arr
```
